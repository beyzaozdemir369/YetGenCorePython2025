# Problem 1

Şimdi beraber kullanıcılardan aldığımız verileri işleyerek, onlara hem çıktı verecek, hem de aldığımız ve hesapladığımız verileri ileride kullanabilmek adına kaydedeceğiz.


 **Bu bir veri giriş platformu olacak**


 * kullanıcıya kaç giriş yapacağını soracağız
* kullanıcıdan isim-soyisim, yaş, boy, kilo bilgilerini alarak VKİ hesaplayacağız
* ardından bunları sınıflara göre kaydedeceğiz bir dosyaya kaydedeceğiz


Sınıflar ise şu şekilde: 
* 18, 5 kg/m.'nin altında olanlar: Zayıf
* 18.5 – 24, 9 kg/m. arasında olanlar: Normal kilolu
* 25 – 29, 9 kg/m. arasında olanlar: Fazla kilolu
* 30 – 39, 9 kg/m. arasında olanlar: Obez
* 40 kg/m.'nin üzerinde olanlar: İleri derecede obez (morbid obez), olarak görülür.

```python
def calculate_bmi_classification(name, surname, age, height, weight):
    height_m= height/ 100 
    bmi = weight / (height_m ** 2)

    if bmi < 18.5:
        classification = "Zayıf"
    elif 18.5 <= bmi <= 24.9:
        classification = "Normal kilolu"
    elif 25 <= bmi <= 29.9:
        classification = "Fazla kilolu"
    elif 30 <= bmi <= 39.9:
        classification = "Obez"
    else:
        classification = "İleri derecede obez (morbid obez)"
    
    return {
        "İsim": name,
        "Soyisim": surname,
        "Yaş": age,
        "Boy(cm)": height,
        "Kilo(kg)": weight,
        "VKİ": round(bmi, 2),
        "Sınıf": classification
    }

def main():
    kisi_sayisi = int(input("Kaç kişi için bilgi gireceksiniz? "))
    tum_kayitlar = []

    for i in range(kisi_sayisi):
        print(f"\n{i+1}. kişi için bilgiler:")
        ad = input("İsim: ")
        soyad = input("Soyisim: ")
        yas = int(input("Yaş: "))
        boy = float(input("Boy (cm): "))
        kilo = float(input("Kilo (kg): "))

        kayit = calculate_bmi_classification(ad, soyad, yas, boy, kilo)
        tum_kayitlar.append(kayit)

    with open("vki_kayitlari.txt", "w", encoding="utf-8") as dosya:
        for kisi in tum_kayitlar:
            for anahtar, deger in kisi.items():
                dosya.write(f"{anahtar}: {deger}\n")
            dosya.write("-" * 30 + "\n")
    
    print("\nTüm veriler 'vki_kayitlari.txt' dosyasına kaydedildi.")

main()
```

# Problem 2
**YetBar - Smoothie Bank**
---
YetBar spor salonunun girişinde bulunan, insansız bir sağlıklı içecek otomatıdır. Kullanıcı çeşitli girdiler smoothie içeceğine koyacağı ürünleri belirleyebilir, fiyat hesaplaması yaptırabilir ya da sadece menüyü görüntüleyebilir.

**Projenin Amacı:**

*   Birbirini etkileyen ve bir kod içerisinde birden fazla defa başvurulacak fonksiyonların oluşturulup, "fonksiyonel" proglamlamak.
*   Büyük programları yazarken, yazılımcı olarak yapabileceğimiz hatalardan kaçınmak için kontrol yöntemleri geliştirmek.
*   Kullanıcı inputlarını ve yaptığımız hesapları, raporlamak/loglamak amacıyla kaydetmek.

**Programdan Beklenen Çıktılar**
* Program kullanıcıya temelde 2 farklı işlem yapma imkanı sağlar.
> 1. Menüyü görüntüleme
> 3. Satın alma
* Satın alınan her içecek tarih, fiyat bilgisi ile kaydedilmeli ve her zaman ulaşılabilir olmalıdır.




```python
import datetime


menu = {
    "Yeşil Detoks": 45,
    "Protein Bombası": 55,
    "Meyve Patlaması": 50,
    "Enerji Dalgası": 60,
    "Kahveli Smoothie": 65
}


siparisler = []

def menu_goster():
    print("\n--- YETBAR MENÜ ---")
    for urun, fiyat in menu.items():
        print(f"{urun}: {fiyat} TL")

def satin_al():
    menu_goster()
    secim = input("\nSatın almak istediğiniz içeceğin adını girin: ")
    if secim in menu:
        fiyat = menu[secim]
        tarih = datetime.datetime.now().strftime("%Y-%m-%d %H:%M:%S")
        siparis = {
            "İçecek": secim,
            "Fiyat": fiyat,
            "Tarih": tarih
        }
        siparisler.append(siparis)
        print(f"\n{secim} başarıyla {fiyat} TL'ye satın alındı.")
    else:
        print("\nHatalı giriş! Lütfen menüden bir içecek adı girin.")

def siparisleri_kaydet():
    with open("siparis_kayitlari.txt", "w", encoding="utf-8") as dosya:
        for siparis in siparisler:
            dosya.write(f"İçecek: {siparis['İçecek']}\n")
            dosya.write(f"Fiyat: {siparis['Fiyat']} TL\n")
            dosya.write(f"Tarih: {siparis['Tarih']}\n")
            dosya.write("-" * 30 + "\n")
    print("\nSiparişler başarıyla 'siparis_kayitlari.txt' dosyasına kaydedildi.")

def programi_baslat():
    while True:
        print("\n--- YETBAR SMOOTHIE OTOMATI ---")
        print("1. Menüyü Görüntüle")
        print("2. Satın Al")
        print("3. Çıkış")

        secim = input("Seçiminiz (1/2/3): ")

        if secim == "1":
            menu_goster()
        elif secim == "2":
            satin_al()
        elif secim == "3":
            siparisleri_kaydet()
            print("Programdan çıkılıyor...")
            break
        else:
            print("Geçersiz seçim, lütfen tekrar deneyin.")


programi_baslat()
```
